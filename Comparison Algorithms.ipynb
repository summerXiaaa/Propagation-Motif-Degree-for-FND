```python
import networkx as nx
import os
import time
import pandas as pd

class motif:
    """
    树状有向网络模体类，模体种类数是
    """
    def __init__(self):
        self.M5 = list()
        self.M5.append(nx.from_edgelist([(1, 2), (1, 3), (1, 4), (1, 5)], create_using=nx.DiGraph()))  # M5_1
        self.M5.append(nx.from_edgelist([(1, 2), (2, 3), (3, 4), (4, 5)], create_using=nx.DiGraph()))  # M5_2

def Isomorph(g1, g2) -> bool:
    return nx.algorithms.isomorphism.is_isomorphic(g1, g2)

def classify(g, Motif) -> int:
    n = len(Motif)
    for i in range(n):
        if Isomorph(g, Motif[i]):
            return int(i)
    return -1

def search(deep, k, G, sub_node, Motif, result, l, node_num, node_list):
    if deep == k:
        H = G.subgraph(sub_node)
        a = classify(H, Motif)
        if a != -1:
            result[a] += 1
    else:
        for j in range(l, node_num):
            node = node_list[j]
            sub_node.append(node)
            search(deep + 1, k, G, sub_node, Motif, result, j + 1, node_num, node_list)
            sub_node.pop()

def ESU(G, Motif):
    node_list = list(G.nodes)
    node_num = len(node_list)
    result = [0] * len(Motif)
    k = len(list(Motif[0].nodes))
    sub_node = []
    search(0, k, G, sub_node, Motif, result, 0, node_num, node_list)
    return result

if __name__ == "__main__":
    total_starttime = time.time()
    Moti = motif()
    
    
    # Read and process the target file
    print(f"Processing {file_path} ...")
    
    # Read the CSV file and create the graph
    df = pd.read_csv("t3617138864185835.csv", encoding='ANSI')
    G = nx.from_pandas_edgelist(df, source='parent', target='mid', create_using=nx.DiGraph())
    
    # Calculate the time to count the first M5 motif (M5_1)
    starttime = time.time()
    count_M5_1 = ESU(G, [Moti.M5[0]])
    endtime = time.time()
    print(f"M5_1 motif counts: {count_M5_1}, Time is: {endtime - starttime} seconds")
    
    
    starttime = time.time()
    count_M5_2 = ESU(G, [Moti.M5[1]])
    endtime = time.time()
    print(f"M5_2 motif counts: {count_M5_2}, Time is: {endtime - starttime} seconds")
    
    total_endtime = time.time()
    print('Total motif time for the target file:', total_endtime - total_starttime, 'seconds')
```
